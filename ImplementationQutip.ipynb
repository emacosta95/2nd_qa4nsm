```python
from ManyBodySystemQutip.qutip_class import SpinOperator
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_qubits=6
data_onebody=np.load('data/matrix_elements_h_eff_2body/one_body_nn_sd.npz')
keys=data_onebody['keys']
values=data_onebody['values']

t_onebody={}

for a,key in enumerate(keys):
    i,j=key
    t_onebody[tuple(key)]=values[a]
    if i==j:
        print(values[a])
print(t_onebody)
```

```python
from qutip import basis, tensor

nqubits = 6  # or however many qubits you have
labels = [format(i, f"0{nqubits}b") for i in range(2**nqubits)]
```

```python
print(labels)
```

```python
index=[]
index_diagonal=[]
index_diagonal_2=[]
values_diagonal=[]
values=[]
for a in range(n_qubits):
    for b in range(n_qubits):
        if t_onebody[a,b]!=0 and a!=b:
            index.append(('x',a,'x',b))
            values.append(0.25*t_onebody[(a,b)])
        elif t_onebody[a,b]!=0 and a==b:
            values_diagonal.append(0.5*t_onebody[(a,b)])
            index_diagonal.append(('z',a))
            index_diagonal_2.append(('id',a))
                
```

```python

hamiltonian_couplings=SpinOperator(index=index,coupling=values,size=n_qubits)
hamiltonian_diagonal_part1=SpinOperator(index=index_diagonal,coupling=values_diagonal,size=n_qubits)
hamiltonian_diagonal_part2=SpinOperator(index=index_diagonal_2,coupling=values_diagonal,size=n_qubits)

hamiltonian_q=hamiltonian_couplings.qutip_op+hamiltonian_diagonal_part1.qutip_op+hamiltonian_diagonal_part2.qutip_op
```
